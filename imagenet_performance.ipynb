```python
import matplotlib.pyplot as plt
import numpy as np
barWidth = 0.5
x = np.arange(7)
error1 = [28.2, 25.8]
error2 = [16.4, 11.7, 7.3, 6.7, 3.57]
error = error1+error2
p1 = [0,1]
p2 = [2,3, 4, 5, 6]
label = ['28.2', '25.8', '16.4', '11.7', '7.3', '6.7', '3.57']
l_net = ['0', '1', 'AlexNet', 'ZF', 'VGG', 'GooLeNet', 'ResNet']
layer = [1,1,8,8,16,19,29]
l_layer = ['shallow','shallow','8 layers','8 layers','19 layers','22 layers','152 layers']
threshold = 5.0
fig, ax = plt.subplots(figsize=(13, 7))
ax.tick_params(axis="x", labelsize=13)
ax.tick_params(axis="y", labelsize=14)

plt.bar(p1, error1, width = barWidth, color =(0.3,0.9,0.4,0.6),label='Shallow model error rate')
plt.bar(p2, error2, width = barWidth, color ='g', label='Deep model error rate')
# ax.plot([0., 7.5], [threshold, threshold], "r--")
plt.axhline(y=threshold,linewidth=2, linestyle='dashed', color='r', label='Human error rate')
plt.plot(x, layer, 'b-^', label='Network depth')
# plt.plot(x, layer, 'b^')
# plt.title('ImageNet Classification Performance over Time', fontsize=18)
plt.ylabel('Error Rate (%)', fontsize=18)
plt.xticks(x, ('2010', '2011', '2012 AlexNet', '2013 ZF', '2014 VGG', '2014 GooLeNet', '2015 ResNet'))
# Text on the top of each barplot
for i in range(len(x)):
    plt.text(x = x[i]-0.15, y = error[i]+0.3, s = label[i], size = 14)
    if i==3 or i==5:
        plt.text(x = x[i]-0.2, y = layer[i]-1.5, s = l_layer[i], size = 14)
    elif i== len(x)-1:
        plt.text(x = x[i]-0.8, y = layer[i]-0.3, s = l_layer[i], size = 14)
    else:
        plt.text(x = x[i]-0.4, y = layer[i]+0.99, s = l_layer[i], size = 14)
# for i in range(len(x)):
#     if i>1:
#         plt.text(x = x[i]-0.2, y = error[i]+1.6, s = l_net[i], size = 14)
plt.legend(prop={'size': 14})

plt.show()
```
